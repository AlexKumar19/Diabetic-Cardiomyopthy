```python
def process_molecules(smilelist):
    for i in range(len(smilelist)):
        mol = Chem.MolFromSmiles(smilelist[i])
        mol = Chem.AddHs(mol, addCoords=True)
        AllChem.EmbedMolecule(mol)
        AllChem.MMFFOptimizeMolecule(mol)
        AllChem.ComputeGasteigerCharges(mol)
        
        file_path_to_pdb = f"/home/vishnuaa77/vscode/Docking/PDB_Files_of_the_Inhibitors/output{i}.pdb"
        file_path_to_pdbqt = f"/home/vishnuaa77/vscode/Docking/PDBQT Files of the Inhibitors/output{i}.pdbqt"
        
        pdb_data = Chem.rdmolfiles.MolToPDBBlock(mol)
        with open(file_path_to_pdb, "w") as pdb_file:
          pdb_file.write(pdb_data)
        
        mol2 = next(pybel.readfile("pdb", file_path_to_pdb))
        mol2.write("pdbqt", file_path_to_pdbqt, overwrite=True)
```

```python
smiles = smilelist[1]

# Create a molecule object from the SMILES string
mol = Chem.MolFromSmiles(smiles)

# Generate 3D coordinates
mol = Chem.AddHs(mol, addCoords=True)
AllChem.EmbedMolecule(mol)
AllChem.MMFFOptimizeMolecule(mol)

# Assign the partial charges
AllChem.ComputeGasteigerCharges(mol)

# Save the molecule in PDB format
w = Chem.PDBWriter("prepared.pdb")
w.write(mol)
w.close()

# Open the PDB file for writing
with open("prepared.pdb", "a") as pdb_file:
    # Iterate over the atoms
    for i, atom in enumerate(mol.GetAtoms()):
        x,y,z = mol.GetConformer().GetAtomPosition(i)
        pdb_file.write("ATOM  {:5d} {:4s} UNL     1    {:8.3f}{:8.3f}{:8.3f}  1.00  {:6.3f}           {:2s}\n".format(
            i+1, atom.GetSymbol(), x, y, z, atom.GetDoubleProp("_GasteigerCharge"), atom.GetSymbol()
        ))

mol2 = next(pybel.readfile("pdb", '/home/vishnuaa77/vscode/Docking/prepared.pdb'))
mol2.write("pdbqt", file_path_to_pdbqt, overwrite=True)
```
