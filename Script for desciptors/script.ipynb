```python
import padelpy
import pandas as pd
import os
def smilestopadel(smiles_list):
    descriptors_list = []
    for i, smiles in enumerate(smiles_list):
        try:
            descriptors = padelpy.from_smiles(smiles)
            descriptors_list.append(pd.DataFrame([descriptors]))
        except:
            print(f"Error calculating descriptors for SMILES {i+1}: {smiles}")
            continue

    result_df = pd.concat(descriptors_list, axis=0, ignore_index=True)

    return result_df
```

```python
import pandas as pd

df = pd.read_csv('Drugbank.csv')


smiles_list = df['SMILES'].tolist()

dataframe = smilestopadel(smiles_list)
```

```python
dataframe.to_csv('FDADescriptsDrugBankForRealThisTime.csv')
```
